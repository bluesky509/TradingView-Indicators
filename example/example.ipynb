```python
import pandas as pd
import tradingview_indicators as ta
```

```python
df = pd.read_csv("BTCUSDT_1d_spot.csv", index_col=0).drop(columns='open_time_ms')
df
```

```python
source = df["close"].copy()
df["RSI"] = ta.RSI(source, 14)
df["MACD"] = ta.MACD(source, 12, 26, 9).get_histogram
df['EMA'] = ta.ema(source, 14)
df["SMA"] = ta.sma(source, 14)
df["RMA"] = ta.rma(source, 14)
df["TEMA"] = ta.sema(source, 14, 3)

dmi = ta.DMI(df, "close")

df["ADX"] = dmi.adx()[0]
df["DI+"] = dmi.adx()[1]
df["DI-"] = dmi.adx()[2]
df["DI_Delta"] = dmi.di_difference()[0]
df["DI_Ratio"] = dmi.di_difference()[1]
df
```
